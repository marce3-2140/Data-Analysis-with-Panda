# Estadisticas resumidas

con df.describe() se obtiene las siguientes estadisticas resumidas:

````python
df.describe()
````

# .round()para fijar cunatos decimales se quieren en el resultado.

*Si prefieren tener menos precision en sus resultados y eliminar la notacion cientifica, pueden fijar cuantos lugares ocupar despues del decimal*

````python
df.describe().round(2)
````

# Incluir todas ('all') las variables en el resultado de .describe()

*Por defecto solo las variables numericas estan en la salida, para incluir todas debe usar:* **df.describe(include='all') tambien proporcionara el recuento de las columnas no numericas.**

````python
df.describe(include='all').round(2)
````
