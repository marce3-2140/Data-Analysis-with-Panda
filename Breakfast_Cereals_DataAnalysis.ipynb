# **Breakfast Cereals: Healthy or Unhealthy?**

> **DataFrame context:**

Cold cereals are an easy, convenient food. Many boast impressive health claims or try to promote the latest nutrition trend. But you may wonder whether these cereals are as healthy as they claim to be.

Breakfast cereal is made from processed grains and often fortified with vitamins and minerals. It is commonly eaten with milk, yogurt, fruit, or nuts.

Here is how breakfast cereals are typically made (3):

1. Processing. The grains are usually processed into fine flour and cooked.
2. Mixing. The flour is then mixed with ingredients like sugar, cocoa, and water.
3. Extrusion. Many breakfast cereals are produced via extrusion, a high-temperature process that uses a machine to shape the cereal.
4. Drying. Next, the cereal is dried.
5. Shaping. Finally, the cereal is shaped into forms, such as balls, stars, loops or rectangles.

###III. **Data Visualization**

```python
#correlation matrix
corr = df.corr()
sns.heatmap(corr, cmap = 'Greens')
```

```python
#Relation between the content of milligrams of sodium and manufacturer company
df.boxplot(column= 'milligrams of sodium', by='Manufacturer')
plt.xticks(rotation = 70);
```

1. In the above graph we can see, that the Nabisco manufacture has the least amount of milligrams of sodium in its cereals, followed by Quakers Oats.

2. The rest of the manufactures work with much higher amounts of milligrams of sodium and they are similar to each other.

```python
##Relation between the content of grams of protein and manufacturer company
df.boxplot(column= 'grams of protein', by='Manufacturer')
plt.xticks(rotation = 70);
```

1. In the above graph we can see, similarity between manufactures in terms of grams of protein in cereals.

```python
#Relation between the content of vitamins and minerals (% of FDA recomendation) and manufacturer company
df.boxplot(column= 'vitamins and minerals (% of FDA recommendation)', by='Manufacturer')
plt.xticks(rotation = 70);
```

1. In general, the previous graph shows that the manufactures provide the same content of vitamins and minerals, except Quaker Oats and Nabisco.

2. It is necessary to review and compare their relationship with the FDA recommendations.

```python
#Definimos las variables a graficar, el ipo de grafico y colores
g = sns.catplot(
    data=df, kind="bar",
    x='Manufacturer', y="grams of sugars", hue="type",
    errorbar="sd", palette="dark", alpha=.6, height=6
    )

#definimos las leyendas y el titulo.
g.despine(left=True)
g.set_axis_labels("", "Grams of Sugars")
g.legend.set_title("")
plt.title('Relation between the Content of Grams of Sugar and Cereals Type')
plt.xticks(rotation=90);
```

The Nabisco manufacturer has de lowest levels of sugar content in its cereals, the other have a similar content and the Post manufacturer contains the highest levels of this.
